```python
fraud_coun=df["Class"].value_counts()
```

```python
fraud_rate=100*fraud_coun/df.shape[0]
```

```python
fraud_data = pd.concat([fraud_coun,fraud_rate],axis=1).reset_index()
fraud_data.columns=['class','count','per']
fraud_data
```

```python
df_fraud= df[df['Class']==1]
df_not_fraud= df[df['Class']==0]
df_not_fraud_sampled=df_not_fraud.sample(df_fraud.shape[0],replace=False,random_state=101)
df_balance= pd.concat([df_not_fraud_sampled,df_fraud],axis=0).sample(frac=1,replace=False,random_state=101).reset_index().drop('index',axis=1)
df_balance                     
```

```python
fraud_count= df_balance["Class"].value_counts()
fraud_rate = 100*fraud_count/df_balance.shape[0]
fraud_data = pd.concat([fraud_count,fraud_rate], axis=1).reset_index()
fraud_data.columns = ['Class', 'Count','Percentage']
fraud_data
```

```python
X_train, X_test, y_train,y_test = train_test_split(df_balance.drop('Class', axis=1), df_balance['Class'], test_size=0.2, random_state=101)
X_train.shape,y_train.shape
```

fit a random forest model

```python
randomForestModel =Pipeline([ ('scaler', StandardScaler ()),('classifier', RandomForestClassifier ())
                            ])
randomForestModel.fit(X_train, y_train)
```

```python
y_pred_logis = randomForestModel.predict(X_test)
y_pred_logis
```

```python
cr =classification_report(y_test, y_pred_logis)
print(cr)
```
